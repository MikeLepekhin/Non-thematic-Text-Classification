```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from natasha import Doc, Segmenter, MorphVocab, NewsEmbedding, NewsMorphTagger

segmenter = Segmenter()
morph_vocab = MorphVocab()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
```

```python
def lemmatize_text(text):
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    result = []
    
    for token in doc.tokens:
        token.lemmatize(morph_vocab)
        result.append(token.lemma)
    return ' '.join(result)
```

```python
ru_train = pd.read_csv('/home/mlepekhin/data/ru_train')
ru_test = pd.read_csv('/home/mlepekhin/data/ru_test')
```

```python
lemmatized_test = []
for value in ru_test.text.values:
    lemmatized_test.append(lemmatize_text(value))
```

```python
lemmatized_train = []
for value in ru_train.text.values:
    lemmatized_train.append(lemmatize_text(value))
```

```python
ru_test['text'] = lemmatized_test
ru_train['text'] = lemmatized_train
```
