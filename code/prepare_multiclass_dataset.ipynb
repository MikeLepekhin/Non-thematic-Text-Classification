```python
all_en = pd.read_csv('en.csv', sep='\t')
print(all_en.shape)
all_en.head()
```

It looks like the discussion texts are labeled with A1.

```python
all_ru = pd.read_csv('ru.csv', sep='\t')
print(all_ru.shape)
all_ru.head()
```

```python
def make_df(file, class_df):
  '''
  Function for forming a dataframe

  Parameters.
  1) file - address of the file with texts,
  2) class_df - pd.DataFrame with class labels for the texts. 
  '''

  result_df = pd.DataFrame()

  with open(file, 'r') as fin:
    for line_id, line in enumerate(fin):
      for category in class_df.columns[1:]:
        if class_df.iloc[line_id][category] == 2:
          result_df = result_df.append({'text': line, 'target': category}, ignore_index=True)
          break
  return result_df
```

```python
df_en = make_df('en.ol', all_en)
df_ru = make_df('ru.ol', all_ru)
```

One of the experiments to be carried out is comparison of multilingual models with monolingual ones. Let us make a multilingual dataset.

```python
plt.hist(df_ru['target'])
plt.title('Class Distribution in The Russian Dataset', fontsize=18)
plt.show()
```

```python
plt.hist(df_multi['target'])
plt.title('Class Distribution in The Multilingual Dataset', fontsize=16)
plt.show()
```

As we see, multilingual data is quite balanced when the both monolingual datasets are not.

Therefore, since identification of both classes is equally important, during training the monolingual models we will add the texts of the minor class twice into the training data.

Finally, we create data sets that will be used for training models.

```python
df_en = df_en.sample(frac=1)
df_en_train = df_en.iloc[:int(0.75*df_en.values.shape[0]), :]
df_en_test = df_en.iloc[int(0.75*df_en.values.shape[0]):, :]
df_en_train.to_csv('/content/drive/My Drive/abbyy/en_train')
df_en_test.to_csv('/content/drive/My Drive/abbyy/en_test')

df_ru = df_ru.sample(frac=1)
df_ru_train = df_ru.iloc[:int(0.75*df_ru.values.shape[0]), :]
df_ru_test = df_ru.iloc[int(0.75*df_ru.values.shape[0]):, :]
df_ru_train.to_csv('/content/drive/My Drive/abbyy/ru_train')
df_ru_test.to_csv('/content/drive/My Drive/abbyy/ru_test')
```

```python
df_multi_train = pd.concat([df_ru_train, df_en_train])
df_multi_train = df_multi_train.sample(frac=1)
df_multi_test = pd.concat([df_ru_test, df_en_test])
df_multi_train.to_csv('/content/drive/My Drive/abbyy/multi_train')
df_multi_test.to_csv('/content/drive/My Drive/abbyy/multi_test')
df_multi_train.head()
```
